```python
from matplotlib import pyplot as plt
avg_x = 0
avg_y = 0
avg_xy = 0
avg_x2 = 0
avg_xx = 0
x = [1,2,4,7,5,2,6,7,4]
y = [2,1,3,6,4,5,9,8,10]

plt.xlabel("x")
plt.ylabel("y")
plt.axis([0,10,0,10])

plt.scatter(x,y)

for i in range(0,len(x)):
    label = "(" + str(x[i])+","+str(y[i])+")"
    plt.annotate(label,xy = (x[i],y[i]),xytext = (5,-5),textcoords = 'offset points')
    
for i in range(0,len(x)):
    avg_x += x[i] / len(x)
    avg_y += y[i] /len(x)
    avg_xy += x[i]*y[i] /len(x)
    avg_x2 = avg_x ** 2
    avg_xx += (x[i]*x[i]) / len(x)
    

m = ((avg_x * avg_y) - (avg_xy)) / (avg_x2 - avg_xx)
b = avg_y -(m*avg_x)

x4line = []
y4line = []

for i in range(0,11):
    x4line.append(i)
    y4line.append(m*i+b)

print(m,b)
plt.plot(x4line,y4line, 'k-', label = 'y = %s'% str(m * i + b) , linewidth = 1)
plt.legend(loc = 9)
plt.grid()
```
