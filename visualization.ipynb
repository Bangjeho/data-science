```python
from matplotlib import pyplot as plt

movies = ["Annie Hall","Ben-Hur","Casablanca","Gandhi","West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
colors = [(1,0.4,0.6,0.6),(1.0,0.3,1.0,0.6),(0.4,0.5,1.0,0.6),
          (0,1,0,0.6),(1.0,0.3,1.0,0.6)]
edgecolors = ['blue','black','cyan','red','green']

plt.bar(movies,num_oscars, color = colors, edgecolor = edgecolors, linewidth = 3 )
plt.ylabel("#of Academy Awards")
plt.title("My Favorite Moives")

#plt.show()

# download
plt.savefig("download.png")
```

```python
from matplotlib import pyplot as plt

x = [1,2,4,8,16,32,64,128,256]

# xs = [i for i, _ in enumerate(x)]
xs = []
for i in range(0,len(x)):
    xs.append((i+1)*100)

plt.plot(xs,x, 'g-', label = 'variance')

plt.legend(loc = 9)
plt.show()
```

```python
from matplotlib import pyplot as plt

x = [1,2,4,8,16,32,64,128,256]
y = []
for i in range(len(x)-1,-1,-1):
    y.append(x[i])

z= []
for i in range(0,len(x)):
    z.append(x[i] + y[i])
xs = []
for i in range(0,len(x)):
    xs.append((i+1)*100)

plt.plot(xs,x, 'g-', label = 'variance')
plt.plot(xs,y,'r-',label = 'bias^2')
plt.plot(xs,z,'b-',label = 'total error')
plt.legend(loc = 9)
plt.show()
```

```python
from matplotlib import pyplot as plt

x = [70,65,72,63,71,64,60,64,67]
y = [175,170,205,120,220,130,105,145,190]

plt.scatter(x,y)

for i in range(0, len(x)):
    label = "(" + str(x[i])+","+str(y[i])+")"
    plt.annotate(label,xy = (x[i],y[i]),xytext = (5,-5),textcoords = 'offset points')
plt.show()
```
