```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("simpsons.csv")
data['gender'] = data['gender'].map({'M':0,'F':1})
data.rename(columns = {'Unnamed: 0':'Name'}, inplace=True)
```

```python
import math
def find_entropy(gender):
    global entropy
    
    genders = gender.groupby('gender').count()
    
    if genders.empty :
        male = 100
        female = 100
    else:
        if genders.index[0] == 0 and len(genders.index) == 2:
            male = genders.values[0][0]
            female = genders.values[1][0]
        else:        
            if genders.index[0] == 0:
                male = genders.values[0][0]
                female = 0.0000000001
            
            elif genders.index[0] == 1:
                male = 0.0000000001
                female = genders.values[0][0]
                
        total = male + female
        left_node_entropy = (male / total) * (-1 * (math.log2(male/total)))
        right_node_entropy = (female / total) * (-1 * (math.log2(female/total)))

        entropy = left_node_entropy + right_node_entropy

    return entropy
```

```python
#최적의 머리 길이 구하기위한 케이스 2가지
list_case1 = []
list_case2 = []

def hair_length_case1(data_values):
    x = data[data.hair_length >= data_values]
    y = data[data.hair_length < data_values]
    
    x_entropy = find_entropy(x)
    y_entropy = find_entropy(y)
    
    entropy = x_entropy + y_entropy
    
    list_case1.append(entropy)

def hair_length_case2(data_values):
    x = data[data.hair_length > data_values]
    y = data[data.hair_length <= data_values]
    
    x_entropy = find_entropy(x)
    y_entropy = find_entropy(y)
    
    entropy = x_entropy + y_entropy
    
    list_case2.append(entropy)
```

```python
#최적의 머리 길이를 찾는 러닝
def hair_length_learning():
    for i in range(0,len(data)):
        hair_length_case1(data['hair_length'][i])
    
        hair_length_case2(data['hair_length'][i])
    
    hair_list_case1 = np.array(list_case1)

    hair_list_case2 = np.array(list_case2)

    if hair_list_case1.min() >= hair_list_case2.min():
        hair_best_entropy = hair_list_case2.min()

    elif hair_list_case1.min() < hair_list_case2.min():
        hair_best_entropy = hair_list_case1.min()
        
    return hair_best_entropy
```
