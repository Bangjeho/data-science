```python
import matplotlib.pyplot as plt

def draw_vectors(vecs):
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.xticks(np.arange(-10,10,1))
    plt.yticks(np.arange(-10,10,1))
    plt.grid(color = '#CCCCCC', linestyle = '-', linewidth = 1)
    
    x = vecs[:,0:1].reshape(np.size(vecs,0))
    y = vecs[:,1:2].reshape(np.size(vecs,0))
    
    plt.scatter(x,y,color = 'black', zorder = 2)

def linear_transform(matrix,vecs):
    t_vecs = []
    for vec in vecs:
        t_vec = matrix.dot(vec)
        t_vecs.append(t_vec)
    t_vecs = np.array(t_vecs)
    
    return t_vecs
```

```python
import numpy as np

vectors = np.array([[2,1],[3,1],[4,1],
                   [3,2],[3,3],[3,4],
                   [3,5],[3,6],[2,5]])
draw_vectors(vectors)
```

```python
basis_vectors = np.array([[0,-1],
                        [1,0]])
t_vectors = linear_transform(basis_vectors,vectors)
draw_vectors(t_vectors)
```

```python
basis_vectors = np.array([[1,1],
                        [-1,1]])
t_vectors = linear_transform(basis_vectors,vectors)
draw_vectors(t_vectors)
```

```python
basis_vectors = np.array([[0,0],
                        [0,0]])
t_vectors = linear_transform(basis_vectors,vectors)
draw_vectors(t_vectors)
```

```python
basis_vectors = np.array([[-1,0],
                        [0,1]])
t_vectors = linear_transform(basis_vectors,vectors)
draw_vectors(t_vectors)
```

```python
basis_vectors = np.array([[1,1],
                        [0,1]])
t_vectors = linear_transform(basis_vectors,vectors)
draw_vectors(t_vectors)
```
