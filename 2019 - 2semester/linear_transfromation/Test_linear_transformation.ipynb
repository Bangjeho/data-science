```python
def linear_transform(new_basis, vector):
    new_vector = []
    new_vector.append(vector[0]*new_basis[0][0] + vector[1]*new_basis[1][0])
    new_vector.append(vector[0]*new_basis[0][1] + vector[1]*new_basis[1][1])
    return new_vector
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw(vectors):
    major_ticks = np.arange(-10, 20, 1)

    fig, ax = plt.subplots()
    ax.set_xticks(major_ticks)
    ax.set_yticks(major_ticks)
    ax.grid()

    plt.xlim(-10, 20)
    plt.ylim(-10, 20)
    x = []
    y = []
    for vector in vectors:
        x.append(vector[0])
        y.append(vector[1])

    plt.scatter(x, y)
    plt.show()
```

```python
vectors = [[2, 1], [3, 1], [4, 1], [3, 2], [3, 3], [3, 4], [3, 5], [3, 6], [2, 5]]
draw(vectors)
```

```python
vectors = [[2, 1], [3, 1], [4, 1], [3, 2], [3, 3], [3, 4], [3, 5], [3, 6], [2, 5]]

new_basis = [[-1, 0], [0, 1]]

for i in range(len(vectors)):
    vectors[i]=linear_transform(new_basis, vectors[i])

draw(vectors)
```

```python
vectors = [[2, 1], [3, 1], [4, 1], [3, 2], [3, 3], [3, 4], [3, 5], [3, 6], [2, 5]]

new_basis = [[1, 0], [0, -1]]

for i in range(len(vectors)):
    vectors[i]=linear_transform(new_basis, vectors[i])

draw(vectors)
```
