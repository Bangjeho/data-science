```python
import tarfile
from six.moves import urllib

url = "http://bike.snu.ac.kr/test/housing.tgz"
path = "housing.tgz"

urllib.request.urlretrieve(url,path)
file = tarfile.open(path)
file.extractall()
file.close()
```

```python
import pandas as pd
data = pd.read_csv("housing.csv")
data.head()
data["ocean_proximity"].value_counts()
```

```python
import matplotlib.pyplot as plt
data.hist(bins = 50, figsize = (20,15))
plt.show()
```

```python
data.plot(kind = 'scatter', x = "longitude", y = "latitude", alpha = 0.3,
         s = data["population"]/100, label = "population",
         c = "median_house_value",cmap = plt.get_cmap("jet"),
         colorbar = True, sharex = False)
```

```python
corr_matrix = data.corr()
corr_matrix["median_house_value"].sort_values(ascending = False)
```

```python
data["rooms_per_household"] = data["total_rooms"] / data["households"]
data["bedrooms_per_room"] = data["total_bedrooms"] / data["total_rooms"]
data['population_per_household'] = data["population"] / data["households"]
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ocean = data[['ocean_proximity']]
encoder = OneHotEncoder()
ocean = encoder.fit_transform(ocean)
ocean.toarray()

ocean = pd.DataFrame(columns = encoder.categories_[0], data = ocean.toarray())
ocean.head(10)
```

```python
value = data[['median_house_value']]
value.head()
```

```python
data.drop(['median_house_value','ocean_proximity'],axis = 1, inplace = True)
```

```python
data = pd.concat([data,ocean,value], axis = 1)
```

```python
data['group'] = pd.cut(x=data['median_house_value'],
                       bins=[0,100000,200000,300000,400000,500001],
                      labels = ['group1','group2','group3','group4','group5'])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits =1,test_size = 0.2, random_state = 42)

for train_index,test_index in split.split(data,data["group"]):
    train = data.loc[train_index]
    test = data.loc[test_index]
```

```python
train.drop('group', axis = 1, inplace = True)
test.drop('group', axis = 1, inplace = True)
```

```python
train_y = train[['median_house_value']]
train.drop('median_house_value',axis = 1, inplace = True)
train_X = train
```

```python
test_y = test[['median_house_value']]
test.drop('median_house_value',axis = 1, inplace = True)
test_X = train
```
